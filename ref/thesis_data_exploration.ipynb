```python
# fig 4.12
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp


df["Compliance"] = 1/df["Total_Stiff"]
#df.rename(columns={"Fat": "Fat Thickness", "Muscle": "Muscle Thickness"})
dfgbl = df.groupby("Location").mean()
my_df = dfgbl[["Fat", "Muscle", "Compliance"]]#.reset_index()
# comp = my_df.pop("Compliance")
# tidy = my_df.melt(id_vars=["Location"]).rename(columns=str.title)
# print(tidy)
# g = sns.catplot(
#     data=tidy, kind="bar",
#     x=tidy.Location, y="Value", hue="Variable", palette="dark", alpha=.6, height=6
# )



fig, ax = plt.subplots(1,2, figsize=(12,6))
coef = np.polyfit(my_df.Fat, my_df.Compliance,1)
poly1d_fn = np.poly1d(coef)
print(poly1d_fn) 
slope, intercept, r_value, p_value, std_err = sp.stats.linregress(my_df.Fat, my_df.Compliance)

ax[0] = sns.scatterplot(data=my_df, x="Fat", y="Compliance", hue=my_df.index,
                        ax=ax[0])
ax[1] = sns.scatterplot(data=my_df, x="Muscle", y="Compliance", hue=my_df.index,
                        ax=ax[1])
ax[0].plot(my_df.Fat, poly1d_fn(my_df.Fat), 'r-')

coef = np.polyfit(my_df.Muscle, my_df.Compliance,1)
poly1d_fn = np.poly1d(coef)
print(poly1d_fn)
slope, intercept, r_value2, p_value, std_err = sp.stats.linregress(my_df.Muscle, my_df.Compliance)

ax[1].plot(my_df.Muscle, poly1d_fn(my_df.Muscle), 'r-')


ax[0].set(xlabel='Fat Thickness (mm)', ylabel='Compliance (mm/MPa)', 
          title=f"R^2={r_value**2:.3f}")
ax[1].set(xlabel='Muscle Thickness (mm)', ylabel='Compliance (mm/MPa)',
          title=f"R^2={r_value2**2:.3f}")


plt.show()
```
